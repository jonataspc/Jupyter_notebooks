# UCL: Graph of the finalists teams 

```python
import networkx as nx
import matplotlib.pyplot as plt

#Data retrieved in https://en.wikipedia.org/wiki/List_of_European_Cup_and_UEFA_Champions_League_finals

teams = [
('Real Madrid','Stade de Reims'),
('Real Madrid','Fiorentina'),
('Real Madrid','Milan'),
('Real Madrid','Stade de Reims'),
('Real Madrid','Eintracht Frankfurt'),
('Benfica','Barcelona'),
('Benfica','Real Madrid'),
('Milan','Benfica'),
('Internazionale','Real Madrid'),
('Internazionale','Benfica'),
('Real Madrid','Partizan'),
('Celtic','Internazionale'),
('Manchester United','Benfica'),
('Milan','Ajax'),
('Feyenoord','Celtic'),
('Ajax','Panathinaikos'),
('Ajax','Internazionale'),
('Ajax','Juventus'),
('Bayern de Munique','Atlético de Madrid'),
('Bayern de Munique','Leeds United'),
('Bayern de Munique','Saint-Étienne'),
('Liverpool','Borussia Mönchengladbach'),
('Liverpool','Club Brugge'),
('Nottinghan Forest','Malmö FF'),
('Nottinghan Forest','Hamburgo'),
('Liverpool','Real Madrid'),
('Aston Villa','Bayern de Munique'),
('Hamburgo','Juventus'),
('Liverpool','Roma'),
('Juventus','Liverpool'),
('Steaua București','Barcelona'),
('Porto','Bayern de Munique'),
('PSV','Benfica'),
('Milan','Steaua București'),
('Milan','Benfica'),
('Estrela Vermelha','Marseille'),
('Barcelona','Sampdoria'),
('Marseille','Milan'),
('Milan','Barcelona'),
('Ajax','Milan'),
('Juventus','Ajax'),
('Borussia Dortmund','Juventus'),
('Real Madrid','Juventus'),
('Manchester United','Bayern de Munique'),
('Real Madrid','Valencia'),
('Bayern de Munique','Valencia'),
('Real Madrid','Bayer Leverkusen'),
('Milan','Juventus'),
('Porto','Monaco'),
('Liverpool','Milan'),
('Barcelona','Arsenal'),
('Milan','Liverpool'),
('Manchester United','Chelsea'),
('Barcelona','Manchester United'),
('Internazionale','Bayern de Munique'),
('Barcelona','Manchester United'),
('Chelsea','Bayern de Munique'),
('Bayern de Munique','Borussia Dortmund'),
('Real Madrid','Atlético de Madrid'),
('Barcelona','Juventus'),
('Real Madrid','Atlético de Madrid'),
('Real Madrid','Juventus'),
('Real Madrid','Liverpool')

]

 

node_size=200
node_color='violet'
node_alpha=0.5
node_text_size=16
edge_color='blue'
edge_alpha=0.8 
edge_tickness=1
edge_text_pos=0.4
text_font='sans-serif'

 
G=nx.Graph()

 
for edge in teams:
   G.add_edge(edge[0], edge[1])

 
graph_pos=nx.spring_layout(G)

 
nx.draw_networkx_nodes(G,graph_pos,node_size=node_size, alpha=node_alpha, node_color=node_color)
nx.draw_networkx_edges(G,graph_pos,width=edge_tickness,alpha=edge_alpha,edge_color=edge_color)
nx.draw_networkx_labels(G, graph_pos,font_size=node_text_size,font_family=text_font)


plt.rcParams['figure.figsize'] = [20, 20]
plt.plot()
 
```
